```python
import pandas as pd 
import matplotlib.pyplot as plt
import os
import sys
sys.path.append('C:\\Users\\longf.DESKTOP-7QSFE46\\GitHub\\A-Stock-Prediction-System-with-GAN-and-DRL')
print(os.path)
```

```python
def drop_and_fill(data):# 去掉重复的列数据 将空值填上数据 输入为股价数据集 输出为去重之后的股价数据集
    data_new = data.T.drop_duplicates(keep='first').T 
    # 去掉重复的数据列 使用转置再转置的方式 非常牛逼
    data_new = data_new.fillna(axis=0, method='ffill')
    return data_new

def get_technical_indicators(data, keys=['close_x_x',], last_days=2691, plot=False):
    # 获取股价技术指标 输入参数为数据集、持续时间和是否绘制图表 输出技术指标 key表示对哪一个指标进行统计分析
    # 7日均线和21日均线
    import stockstats

    tech_total = data['cal_date']
    technical_data = pd.concat([data['cal_date'], data[[key for key in keys]].astype(float)], axis=1)
    for key in keys:
        tech_indicator = pd.DataFrame()
        tech_indicator['date'] = technical_data['cal_date']
        tech_indicator[key + '_tech'] = technical_data[key]
        tech_indicator[key + '_'+'ma7'] = technical_data[key].rolling(window=7).mean()
        tech_indicator[key + '_'+'ma21'] = technical_data[key].rolling(window=21).mean()
        # MACD
        tech_indicator[key + '_'+'26ema'] = pd.DataFrame(technical_data[key]).ewm(span=26).mean()
        tech_indicator[key + '_'+'12ema'] = pd.DataFrame(technical_data[key]).ewm(span=12).mean()
        tech_indicator[key + '_'+'MACD'] = (tech_indicator[key + '_'+'12ema']-tech_indicator[key + '_'+'26ema'])
        # 布林线
        tech_indicator[key + '_'+'20sd'] = technical_data[key].rolling(window=20).std()
        tech_indicator[key + '_'+'upper_band'] = tech_indicator[key + '_'+'ma21'] + (tech_indicator[key + '_'+'20sd']*2)
        tech_indicator[key + '_'+'lower_band'] = tech_indicator[key + '_'+'ma21'] - (tech_indicator[key + '_'+'20sd']*2)
        # 指数移动平均
        tech_indicator[key + '_'+'ema'] = pd.DataFrame(technical_data[key]).ewm(com=0.5).mean()
        # 动量
        tech_indicator[key + '_'+'momentum'] = technical_data[key]-1

        if plot:# 绘制技术指标
            plot_dataset = tech_indicator
            shape_0 = plot_dataset.shape[0]
            xmacd_ = shape_0-last_days
            plot_dataset = plot_dataset.iloc[-last_days:, :]
            x_ = range(0, plot_dataset.shape[0])
            x = list(plot_dataset.index)
            
            plt.figure(figsize=(16, 10), dpi=100)
            # Plot first subplot
            plt.subplot(2, 1, 1)


            plt.plot(plot_dataset[key + '_'+'ma7'], label=key + '_'+'MA 7', color='g',linestyle='--')
            plt.plot(plot_dataset[key + '_'+'tech'], label=key + '_'+'股价技术指标', color='b')
            plt.plot(plot_dataset[key + '_'+'ma21'], label=key + '_'+'MA 21', color='r',linestyle='--')
            plt.plot(plot_dataset[key + '_'+'upper_band'], label=key + '_'+'Upper Band', color='c')
            plt.plot(plot_dataset[key + '_'+'lower_band'], label=key + '_'+'Lower Band', color='c')
            plt.fill_between(plot_dataset[key + '_'+'lower_band'], plot_dataset[key + '_'+'upper_band'], alpha=0.35)
            plt.title(' Technical indicators in {} days .'.format(last_days))
            plt.ylabel(key + '_'+'technical')
            plt.legend()

            # Plot second subplot
            plt.subplot(2, 1, 2)
            plt.title('MACD')
            plt.plot(plot_dataset[key + '_'+'MACD'],label=key + '_'+'MACD', linestyle='-.')
            plt.hlines(15, xmacd_, shape_0, colors='g', linestyles='--')
            plt.hlines(-15, xmacd_, shape_0, colors='g', linestyles='--')
            plt.plot(plot_dataset[key + '_'+'momentum'],label=key + '_'+'Momentum', color='b',linestyle='-')
            plt.legend()
            plt.show()

        # 汇总技术指标
        tech_total = pd.concat([tech_total, tech_indicator.drop(columns='date')], axis=1)

    return tech_total   

data_csv = pd.read_csv('C:\\Users\\longf.DESKTOP-7QSFE46\\GitHub\\A-Stock-Prediction-System-with-GAN-and-DRL\\dataset\\DailyTotal-600690.SH.csv').drop(columns=['Unnamed: 0', 'ts_code_x_x'])
print(data_csv)
cal_date = pd.to_datetime(data_csv['cal_date'], format='%Y%m%d').to_list()
data = drop_and_fill(data_csv)
data = pd.DataFrame(data={col:data[col].tolist() for col in data.columns}, index=cal_date)
keys = ['open', 'high', 'low', 'close_x_x', 'pct_chg', 'amount', 
        'turnover_rate', 'volume_ratio', 'pe', 'pb', 'ps', 'float_share_x', 'total_share_x',
        'buy_sm_amount', 'sell_sm_amount', 'buy_md_amount', 'sell_md_amount',
        'buy_lg_amount', 'sell_lg_amount', 'buy_elg_amount', 'sell_elg_amount']
    # 对以上行情数据进行技术指标分析：
    # 开盘价、收盘价、最高价、最低价、涨跌幅、成交额
    # 换手率、量比、市盈率、市净率、市销率、流通股本、总股本
    # 小、中、大、特大单买入卖出额

    # 获取时间窗口统计技术指标
tech_indicator = get_technical_indicators(data, keys=keys, plot=True)
```
